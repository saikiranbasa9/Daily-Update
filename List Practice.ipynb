## Max/Min Finder
### Given a list of integers, e.g., temperatures = [25, 18, 30, 22, 15], write a program that uses built-in functions (max() and min()) or a loop to find and print the largest and smallest values in the list.

```python
temperature = [25,18,30,22,15]

max1 = max(temperature)
print("the Max of the list is: ",max1)



min1 =min(temperature)
print("the Min of the list is: ",min1)


##Using For Loop

print("Using Loop")


a = temperature[0]

for i in temperature:
    if i > a:
        a = i
print("the Max of the list is: ",a)

b=temperature[0]
for j in temperature:
    if j < b:
        b=j
    
print("the Min of the list is: ",b)    
```

## Duplicate Remover
### Given a list that contains several duplicate elements (e.g., colors = ['red', 'blue', 'red', 'green', 'blue', 'yellow']), write a program to create a new list containing only the unique elements from the original list, preserving the order of first appearance.


```python
colors = ["red","Blue","red","green","blue","yellow"]

unique =[]

for i in colors:
    if i not in unique:
        unique.append(i)
        
        
print(unique)
```

## List Shifter (Slicing)
### Given a list of numbers, e.g., data = [10, 20, 30, 40, 50], use list slicing to "shift" the first two elements to the end of the list. The resulting list should be [30, 40, 50, 10, 20].

```python
data = [10,20,30,40,50]

a = data[:2:]
b = data[2::]


data = b+a
data
```

## List Comprehension Filter
### Given a list of words, use a list comprehension to create a new list that contains only the words that have a length greater than 5 and start with a capital letter.

```python
lst = ["Apple","Banana","Chiku","Dragonfruit","Eggplant","Fig","Grapes"]

a = [word for word in lst if len(word) > 5 and word[0].isupper()]
print(a)

```

## Matrix Row Sums (Nested Lists)
### Given a nested list representing a matrix, e.g., matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]], use a loop (or list comprehension) to calculate and store the sum of each individual row in a new list. The final output list should be [6, 15, 24].

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
b = []
for a in matrix:
    b.append(sum(a))
    
    
print(b)
```
