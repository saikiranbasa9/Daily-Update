## Tuple Unpacking
### Create a tuple containing three items (e.g., coordinates (10, 20, 30)). Use tuple unpacking to assign each value to three separate variables (x, y, z) and print them.

```python
coordinates = (10, 20, 30)

x,y,z = coordinates

print("x = ",x)
print("y = ",y)
print("z = ",z)
```

## Item Indexer
### Given a tuple of fruits, ask the user for a fruit name. Use the index() method to find and print the position (index) of that fruit in the tuple. (Handle the case where the fruit is not found).

```python
fruits = ("apple", "banana", "pineapple", "mango")

user = str(input("Enter a  fruit: ").lower())

index = fruits.index(user)

print(index)
```

## Tuple List Converter
### Given a list of tuples, where each tuple contains a name and an age, e.g., people = [('Alice', 30), ('Bob', 25)]. Use a for loop to print a formatted string for each person, like: "Alice is 30 years old."

```python
people = [('Alice', 30), ('Bob', 25)]

for name,age in people:
    print("the name is:",name,"the age is:",age)
```

## Immutable List Sort
### Given a list of tuples (e.g., scores: [(85, 'A'), (92, 'B'), (78, 'C')]), use the sorted() function with a lambda function to sort the list based on the numerical score (the first element of the tuple), and print the sorted list.

```python
scores = [(85,'A'),(92,'B'),(78,'C'),(95,'A+'),(70,'D')]

sorted_scores = sorted(scores, key=lambda item: item[0])

print(sorted_scores)
```

## Tuple of Lists Editor
### Create a tuple that contains three lists, e.g., data = ([1, 2], [3, 4], [5, 6]). Using indexing, write a line of code to add the number 10 to the second element of the first list within the tuple. This demonstrates the mutability of list contents within an immutable tuple.

```python
data = ([1,2],[3,4],[5,6])

data[0][1] = data[0][1] + 10

print(data)
```

## Unique Tuple List
### Given a list that may contain duplicate tuples, e.g., points = [(1, 2), (3, 4), (1, 2), (5, 6)]. Use the properties of Sets to efficiently remove the duplicates and print the list of unique tuples.

```python
points = [(1,2),(3,4),(1,2),(5,6)]

unique_points = set(points)

unique_points_lst = list(unique_points)


print(unique_points_lst)
```
