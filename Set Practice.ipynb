## Element Checker
### Given a set of prime numbers, primes = {2, 3, 5, 7, 11}, ask the user for a number. Use the in keyword to check and print whether that number is present in the set.

```python
primes = {2, 3, 5, 7, 11, 13}

user = (int(input("enter a number between 1-15: ")))

type(user)


if user in primes:
    print("Its a prime no")
    
else:
    print("its not a prime no")
```

## Set Merger
### Given two sets of names, team_a = {'Alice', 'Bob', 'Charlie'} and team_b = {'Charlie', 'David', 'Eve'}, use the union method (|) to create and print a single set of all unique members.

```python
team_a = {'Alice', 'Bob', 'Charlie'} 
team_b = {'Charlie', 'David', 'Eve'}

team_union = team_a|team_b

print(team_union)
```

## Common Interests
### You have two sets representing hobbies: hobbies_A and hobbies_B. Use the intersection method (&) to find and print the hobbies that both people share.

```python
hobbies_a ={"Drawing", "Music", "singing", "playing"}
hobbies_b ={"Music", "playing"}

hobbies_intersection = hobbies_a & hobbies_b

print(hobbies_intersection)
```

## Unique Differences
### Given two sets of numbers, set1 and set2, use the symmetric difference method (^) to find and print all elements that are in either set, but not in both.

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

set3 = set1^set2

print(set3)
```

## Vowel Word Filter
### Given a list of words, use a loop and set logic to filter the list, keeping only words that contain all five English vowels ('a', 'e', 'i', 'o', 'u').

```python
user1 = str(input("Enter a string to check vowel condition: "))
vowel = set('aeiou')


if vowel.issubset(set(user1.lower())):
    print("All the vowels are present in the string")
else:
    print("All the vowels are not present in the string")
```
