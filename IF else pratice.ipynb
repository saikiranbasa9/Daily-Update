## Simple Calculator with Error Handling
### Problem: Create a program that simulates a basic calculator. It should take two numbers and an operator (+, -, *, /) as input.

```python
a = int(input("Enter number: "))
b = int(input("Enter number: "))
c = str(input("Enter opreator: "))


if c == "+":
    print(a+b)
elif c == "-":
    print(a-b)
elif c == "*":
    print(a*b)
elif c == "/":
    if b == 0 or a == 0:
        print(" Invalid Numbers")
    else:
        print(a/b)
```

## Check for Leap Year
### Problem: Write a program that takes a year as input and determines if it is a leap year.

```python
year = int(input("Enter a year to check leap year status"))

if year%400 == 0 and year%100 == 0 or year%4 == 0:
    print("Its a leap year")
else:
    print("Not a Leap year")
```

## Triangle ClassifierProblem: 
### Write a program that takes the lengths of the three sides of a triangle ($a, b, c$) as input and classifies the triangle into one of three types:

```python
s1 = int(input("Enter side 1 of the triangle "))
s2 = int(input("Enter side 2 of the triangle "))
s3 = int(input("Enter side 3 of the triangle "))

if s1 == 0 or s2 == 0 or s3 ==0:
    print("Side of triangle cannot be zero")

    
if s1+s2 < s3 or s1+s3 < s2 or s2+s3 < s1:
    print("Invalid Sides")
elif s1 == s2 == s3:
    print("Equilateral Triangle")
elif s1 == s2 or s2 == s3 or s1 == s3:
    print("Isosceles triangle")
else:
    print("Scalene Triangle")
```

## Grade Calculator with Weighted Average
## Problem: Create a program that calculates a final course grade based on three weighted components: Assignments (30%), Midterm Exam (30%), and Final Exam (40%).

```python
am = float(input("Enter Assignment Marks: "))
me = float(input("Enter Midtern Exam Marks: "))
fm = float(input("Enter Finalterm Marks: "))

if am < 0 or am > 100:
    print("Invalid Assignment Marks")
elif me < 0 or me > 100:
    print("Invalid Midterm Marks")
elif fm < 0 or fm > 100:
    print("Invalid Finalterm Marks")
    

score = am*.30+me*.30+fm*.40
score


if score >= 90:
    print("A Grade")

elif score >= 80:
    print("B Grade")
    
elif score >= 70:
    print("C Grade")
    
    
elif score >= 60:
    print("D Grade")
else:
    print("F Grade")
```
