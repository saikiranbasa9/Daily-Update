## Word Frequency Counter
### Given a string of text, e.g., "apple orange apple banana orange", use a for loop to count the frequency of each word and store the results in a dictionary (output should be {'apple': 2, 'orange': 2, 'banana': 1})

```python
str1 = "apple orange apple banana orange apple orange banana apple"
count_words = {}
words = str1.split()

for i in words:
    if i in count_words:
        count_words[i] +=1
    else:
        count_words[i] = 1
        
        
print(count_words)
```

## Key-Value Filter
### Given a dictionary of products and prices, e.g., prices = {'milk': 2.5, 'cheese': 4.0, 'bread': 1.5, 'eggs': 3.0}, use a  loop to create a new dictionary containing only items with a price greater than 3.0.

```python
prices = {'milk': 2.5, 'cheese': 4.0, 'bread': 1.5, 'eggs': 3.0}
new_prices ={}

for key,value in prices.items():
    if value > 3:
        new_prices[key] = value
        
print(new_prices)
```

## Nested Data Lookup
### Given a dictionary of countries and their capitals and populations, e.g., data = {'USA': {'capital': 'Washington', 'pop': 330}, 'Canada': {'capital': 'Ottawa', 'pop': 38}}. Write a program to find and print the population of the country with the capital 'Washington'.

```python
data = {    'USA': {'capital': 'Washington', 'pop': 330}, 
            'Canada': {'capital': 'Ottawa', 'pop': 38}}

target_capital = 'Washington'

for country, info in data.items():
    if info['capital'] == target_capital:
        population = info['pop']
        
        
print(f"The population of the country with capital '{target_capital}' is {population} million.")
```

## Dictionary Comprehension Swap
### Using dictionary comprehension, write a single line of code to swap the keys and values of a given dictionary, but only for key-value pairs where the value is an even number.
